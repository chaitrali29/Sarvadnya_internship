```python
from docutils import core
from docutils import nodes
from docutils.parsers.rst import directives, Directive, Parser
from docutils.frontend import OptionParser
from docutils.utils import new_document
import re
import ast

class ManimDirective(Directive):
    has_content = True

    def run(self):
        full_directive = '\n'.join(self.content.data)
        node = nodes.raw('', full_directive, format='manim')
        node.rawsource = full_directive
        return [node]

directives.register_directive('manim', ManimDirective)

# def extract_manim_blocks(source):
#     if source.endswith('.py'):
#         return extract_manim_blocks_from_py(source)
#     elif source.endswith('.rst'):
#         return extract_manim_blocks_from_rst(source)
#     else:
#         raise Exception('Unsupported file type')

"""
def extract_manim_blocks_from_docstring(source):
    # Create a default settings object
    settings = OptionParser(components=(Parser,)).get_default_values()
    
    # Create a new document with these settings
    document = new_document('docstring', settings=settings)
    parser = Parser()

    # Parse the docstring
    parser.parse(source, document)

    # Function to traverse the document tree and extract examples
    def visit_node(node):
        if isinstance(node, nodes.section) and 'examples' in node.astext().lower():
            return node.astext()
        for child in node.children:
            result = visit_node(child)
            if result:
                return result
        return None

    # Extract examples
    examples = visit_node(document)
    # return examples if examples else "No Manim examples found."
    return []
"""

def extract_manim_blocks_from_py(source):
    # Parse the file content into an AST
    tree = ast.parse(source)

    # Function to recursively extract docstrings
    def extract_from_node(node):
        docstrings = []
        if isinstance(node, (ast.FunctionDef, ast.AsyncFunctionDef, ast.ClassDef, ast.Module)):
            docstring = ast.get_docstring(node)
            if docstring:
                docstrings.append(docstring)
        for child in ast.iter_child_nodes(node):
            docstrings.extend(extract_from_node(child))
        return docstrings

    docstrings = extract_from_node(tree)
    manim_blocks = []
    for docstring in docstrings:
        child_manim_blocks = extract_manim_blocks_from_rst(docstring, True)
        manim_blocks.extend(child_manim_blocks)
    return manim_blocks

def extract_manim_blocks_from_rst(source, remove_from_source=False):
    doctree = core.publish_doctree(source)
    manim_blocks = []

    metadata_pattern = re.compile(r'\s*:(\w+):\s*(.*)')

    for node in doctree.traverse(nodes.raw):
        if 'manim::' in node.rawsource or node.attributes['format'] == 'manim':
            all_lines = node.rawsource.split('\n')

            # extract first line as name and rest as remaining lines
            name = all_lines[0]
            lines = all_lines[1:]

            # start_line = node.line
            # end_line = node.line + len(all_lines) - 1
            end_line = node.line
            start_line = node.line - len(all_lines) - 1
# 44-17=27, 22-12=10 
            metadata = [
                {
                    'name': 'name',
                    'value': name
                }
            ]
            code_block_lines = []
            is_metadata = True  # Flag to track whether we are still reading metadata lines
            for line in lines:
                if is_metadata:
                    match = metadata_pattern.match(line)
                    if match:
                        # Extract metadata
                        metadata.append({'name': match.group(1).strip(), 'value': match.group(2).strip()})
                    elif line.strip() and not line.startswith('.. manim::'):
                        # Non-metadata line and not empty, switch to reading code block
                        is_metadata = False
                        code_block_lines.append(line)
                else:
                    # Add line to code block
                    code_block_lines.append(line)

            # Joining the code block lines
            code_block = '\n'.join(code_block_lines).strip()

            # if code block does not contain 'from manim import *' then prepend it
            if 'from manim import' not in code_block:
                code_block = 'from manim import *\n\n' + code_block

            manim_blocks.append({
                'metadata': metadata,
                'code_block': code_block,
                'start_line': start_line,
                'end_line': end_line,
            })


    # remove each child based on start and end line from docstring
    # loop over each line in docstring and only add if no child block contains that line number
    # lines = docstring.split('\n')
    # for line_index, line in enumerate(lines):
    #     for child_block in manim_blocks:
    #         if child_block['start_line'] <= line_index <= child_block['end_line']:
    #             lines[line_index] = ''
    # clean_docstring = '\n'.join(lines)

    if remove_from_source and len(manim_blocks) > 0:
        # filter the lines in source by whether line number is in any child block between start and end line
        lines = source.split('\n')
        filtered_lines = []
        for line_index, line in enumerate(lines):
            is_in_child_block = False
            for child_block in manim_blocks:
                # if child_block['start_line'] <= line_index <= child_block['end_line']:
                # List cases to check:
                # 4. line_index is before start_line then include
                # 1. line_index is start_line then exclude
                # 3. line_index is between start_line and end_line then exclude
                # 2. line_index is end_line then exclude
                # 5. line_index is after end_line then include
                if line_index < child_block['start_line'] or line_index > child_block['end_line']:
                    continue
                else:
                    is_in_child_block = True
                    break
            if not is_in_child_block:
                filtered_lines.append(line)
        clean_source = '\n'.join(filtered_lines)

        # prepend the clean docstring the code_block in each child block
        for child_block in manim_blocks:
            child_block['code_block'] = "\"\"\"\n" + clean_source.strip() + "\n\"\"\"\n\n" + child_block['code_block']

    return manim_blocks


def extract_manim_blocks_from_file(file_path):
    with open(file_path, 'r') as file:
        source = file.read()
    try:
        if file_path.endswith('.py'):
            if 'from manim import *' in source:
                return [
                    {
                        'metadata': [
                            {
                                'name': 'name',
                                'value': file_path
                            }
                        ],
                        'code_block': source
                    }
                ]
            # Parse the docstring
            return extract_manim_blocks_from_py(source)
        elif file_path.endswith('.rst'):
            return extract_manim_blocks_from_rst(source)
        else:
            raise Exception('Unsupported file type')
    except Exception as e:
        print(e)
        return []

def print_manim_blocks(manim_blocks):
    # print count
    print(len(manim_blocks))

    # Displaying the extracted blocks
    for block in manim_blocks:
        print("Metadata:")
        for item in block['metadata']:
            print(item['name'] + ": " + item['value'])
        print("\nCode Block:")
        print(block['code_block'])
        print("\n---\n")
```

```python
# write Manim blocks as JSON to file in train_folder
import json
import os
import hashlib
import re

def extract_scenes_from_manim_file(content):
    # Regular expression to match class definitions that inherit from Scene
    scene_pattern = re.compile(r'class\s+(\w+)\(Scene\):.*?(?=class\s+\w+\(Scene\)|\Z)', re.DOTALL)
    # FIXME: Fails to handle Code example which has Scene inside: https://docs.manim.community/en/stable/reference/manim.mobject.text.code_mobject.Code.html?highlight=fadeinsquare

    # Find all matches and extract scene names and codes
    scenes = []
    for match in scene_pattern.finditer(content):
        scene_name = match.group(1)
        code = match.group(0)
        scenes.append({'sceneName': scene_name, 'code': code})

    return scenes

def write_manim_blocks_to_file(manim_blocks, folder):
    if len(manim_blocks) == 0:
        return []

    if not os.path.exists(folder):
        os.makedirs(folder)

    scenes_in_blocks = []
    for i, block in enumerate(manim_blocks):
        # hash the entire block to get a unique filename
        hash_object = hashlib.md5(block['code_block'].encode())
        filename = hash_object.hexdigest() + '.json'
        with open(folder + '/' + filename, 'w') as outfile:
            json.dump(block, outfile, indent=4)
            py_filename = filename[:-5] + '.py'
            # print(py_filename)

            entire_code = block['code_block']
            scenes = extract_scenes_from_manim_file(entire_code)

            if len(scenes) > 1:
                print("Found {} scenes in {}".format(len(scenes), py_filename))
                # code without any scenes
                code_without_scenes = entire_code
                for scene in scenes:
                    # remove scene from code
                    code_without_scenes = code_without_scenes.replace(scene['code'], '')
                # print(code_without_scenes)

                # code with single scene
                # code_without_scenes + scene's code
                code_for_scenes = []
                for scene in scenes:
                    code_for_scene = code_without_scenes.strip() + "\n\n" + scene['code']
                    code_for_scenes.append({
                        'sceneName': scene['sceneName'],
                        'code': code_for_scene
                    })
                # print(code_for_scenes)
                # loop through code_for_scenes and write to file with folder and py_filename and sceneName
                # for code_for_scene in code_for_scenes:
                for scene_index, code_for_scene in enumerate(code_for_scenes):
                    scene_filename = filename[:-5] + "-scene-" + code_for_scene['sceneName'] + '.py'
                    scenes_in_blocks.append({
                        'sceneIndex': scene_index,
                        'blockIndex': i,
                        'blockFilename': filename,
                        'filename': scene_filename,
                        'folder': folder,
                        'sceneName': code_for_scene['sceneName'],
                        'metadata': block['metadata'],
                        'code': code_for_scene['code'],
                        # 'code_block': block['code_block'],
                    })

                    with open(folder + '/' + scene_filename, 'w') as pyfile:
                        pyfile.write(code_for_scene['code'])

            # Only 1 scene
            else:
                scenes_in_blocks.append({
                    'sceneIndex': 0,
                    'blockIndex': i,
                    'blockFilename': filename,
                    'filename': py_filename,
                    'folder': folder,
                    # 'sceneName': scenes[0]['sceneName'],
                    'sceneName': None,
                    'metadata': block['metadata'],
                    'code': entire_code,
                })
            with open(folder + '/' + py_filename, 'w') as pyfile:
                pyfile.write(entire_code)
    
    return scenes_in_blocks
```

```python
%%manim -qm -v WARNING VMobjectDemo

class VMobjectDemo(Scene):
    def construct(self):
        plane = NumberPlane()
        my_vmobject = VMobject(color=GREEN)
        my_vmobject.points = [
            np.array([-2, -1, 0]),  # start of first curve
            np.array([-3, 1, 0]),
            np.array([0, 3, 0]),
            np.array([1, 3, 0]),  # end of first curve
            np.array([1, 3, 0]),  # start of second curve
            np.array([0, 1, 0]),
            np.array([4, 3, 0]),
            np.array([4, -2, 0]),  # end of second curve
        ]
        handles = [
            Dot(point, color=RED) for point in
            [[-3, 1, 0], [0, 3, 0], [0, 1, 0], [4, 3, 0]]
        ]
        handle_lines = [
            Line(
                my_vmobject.points[ind],
                my_vmobject.points[ind+1],
                color=RED,
                stroke_width=2
            ) for ind in range(0, len(my_vmobject.points), 2)
        ]
        self.add(plane, *handles, *handle_lines, my_vmobject)
```
